# Q1

```python
def count_word():
    try:
        file = input('Enter the file: ')
        with open(file, "r",encoding='utf-8') as f:
            d = dict()
            for line in f:
                line = line.strip()
                line = line.lower()
                line = line.translate(line.maketrans("", "", string.punctuation))
                words = line.split(" ")
                for word in words:
                    if word in d:
                        d[word] += 1
                    else:
                        d[word] = 1
        lg.info('Executed')
        return(list(d.items()))
    except Exception as e:
        lg.error('error')
        lg.exception('exception'+str(e))
```

# Q2

```python
def count_first_words():
    try:
        file = input('Enter the file: ')
        with open(file, "r",encoding='utf-8') as f:
            d = dict()
            for line in f:
                line = line.strip()
                line = line.lower()
                line = line.translate(line.maketrans("", "", string.punctuation))
                words = line.split(" ")
                for word in words:
                    if word in d:
                        d[word] = d[word] + 1
                    else:
                        d[word] = 1
            lst = list(d)
            count = Counter(s[0] for s in  lst)
            final = count.most_common()
            lg.info('executed')
            return final
    except Exception as e:
        lg.error('error')
        lg.exception('exception'+str(e))
```

# Q3

```python
def clean_file():
    try:
        file = input('Enter the file: ')
        with open(file, 'r', encoding='utf-8') as f:
            read_data = f.read()
            punc = '''!()-[]{};:'"\<>./?@#$%^&*_~0123456789=+,'''
            for ele in read_data:
                if ele in punc:
                    read_data = read_data.replace(ele, "")
            vocab_pubmed = ("".join([s for s in read_data.strip().splitlines(True) if s.strip()]))
            with open('vocab_pubmed_without_punctuation.txt','w',encoding='utf-8') as f:
                f.write(vocab_pubmed)
        lg.info('executed')
    except Exception as e:
        lg.error('error')
        lg.exception('exception'+str(e))
```

```python
clean_file() # C:\Users\manim\FSDS\Python\Assignment\vocab.pubmed.txt
```

### convert text files to list.

```python
def convert_list():
    try:
        file = input('Enter the file: ')
        with open(file, "r",encoding='utf-8') as f:
            d = dict()
            for line in f:
                line = line.strip()
                line = line.lower()
                line = line.translate(line.maketrans("", "", string.punctuation))
                words = line.split(" ")
                for word in words:
                    if word in d:
                        d[word] = d[word] + 1
                    else:
                        d[word] = 1
        lst = list(d)
        lg.info('executed')
        return lst
    except Exception as e:
        lg.error('error')
        lg.exception('exception'+str(e))
```

### zipping required text files and storing it in the sqlite database

```python
def zipping():
    try:
        combine = list(zip(text_1,text_2,text_3,text_4,text_5))
        db = sqlite3.connect("sqlite.db")
        c = db.cursor()
        c.execute('create table original(C1 text, C2 text, C3 text, C4 text, C5 text)')
        query = 'insert into original values(?,?,?,?,?)'
        table_rows = combine
        c.executemany(query, table_rows)
        db.commit()
        lg.info('executed')
        return combine
    except Exception as e:
        lg.error('error')
        lg.exception('exception'+str(e))
```

```python
zipping()
```
