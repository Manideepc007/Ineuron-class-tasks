### 3 . create a table called as car with a column name given in dataset description

### 4 . Dump all the data into car table 

```python
import csv
```

```python
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = i[0].split(',')
        query = "INSERT INTO cardataset.car VALUES('{}','{}','{}','{}','{}','{}','{}')".format(str(row_data[0]), str(row_data[1]), str(row_data[2]), str(row_data[3]), str(row_data[4]), str(row_data[5]), str(row_data[6]))
        #query = "INSERT INTO car_db.car VALUES({})".format(str(i[0]))
        cursor.execute(query)
```

### 5 . try to check weather all the data is aviable inside your table or not 

```python
query = "SELECT * FROM cardataset.car"
cursor.execute(query)
cursor.fetchall()
```

### 6 . try to group all the data with COL1 AND count occurences of each and every record based on col1 value

```python
query = "SELECT buying, count(*) FROM cardataset.car GROUP BY buying"
cursor.execute(query)
cursor.fetchall()
```

### 7 . Try to filter a record where col 3 value will be 4 .

```python
query = "SELECT * FROM cardataset.car WHERE doors = '4'"
cursor.execute(query)
cursor.fetchall()
```

### 8.  Try to update a col 3 value with 8 whereever you have value equal to 2 

```python
query = "UPDATE cardataset.car SET doors = '8' WHERE doors = '2'"
cursor.execute(query)
mydb.commit()
```

```python
query = "SELECT doors, count(*) FROM cardataset.car GROUP BY doors"
cursor.execute(query)
cursor.fetchall()
```
